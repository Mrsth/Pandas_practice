```
def myfun(x):
  if x == "four":
    return int(4)
  elif x == "six":
    return int(6)
  elif x == "five":
    return int(5)  
  elif x == "three":
    return int(3) 
  elif x == "twelve":
    return int(12) 
  elif x == "two":
    return int(2)
  elif x == "eight":
    return int(8)  

df2 = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Datasets/Automobile_data.csv")                
df2["Num_of_cylinders"] = df1["num-of-cylinders"].map(lambda x: myfun(x))
df2
```

```
# convertible = 0, hatchback = 1, sedan = 2, wagon = 3, hardtop = 4

def body_map(x):
  if x == "convertible":
    return int(0)
  elif x == "hatchback":
    return int(1)
  elif x == "sedan":
    return int(2)
  elif x == "wagon":
    return int(3)
  elif x == "hardtop":
    return int(4)        

df2["Body_style"] = df2["body-style"].map(lambda x: body_map(x))   
df2 
```

```
df2["engine-type"].unique()
```

```
# dohc = 0, ohcv = 1, ohc = 2, l = 3, rotor = 4, ohcf = 5, dohcv = 6

def engine_type_map(x):
  if x == "dohc":
    return int(0)
  elif x == "ohcv":
    return int(1)
  elif x == "ohc":
    return int(2)
  elif x == "l":
    return int(3)
  elif x == "rotor":
    return int(4)
  elif x == "ohcf":
    return int(5)
  elif x == "dohcv":
    return int(6)

df2["Engine_type"] = df2["engine-type"].map(lambda x: engine_type_map(x))
df2
```

```
# df2.drop(["engine-type"],axis = 1, inplace= True)
df2
```

```
#alfa-romero = 0, audi = 1, bmw = 2, chevrolet = 3, dodge = 4, honda = 5,
#isuzu = 6, jaguar = 7, mazda = 8, mercedes-benz = 9, mitsubishi = 10,
#nissan = 11, porsche = 12, toyota = 13, volkswagen = 14, volvo = 15

def company_map(x):
  if x == "alfa-romero":
    return int(0)
  elif x == "audi":
    return int(1)
  elif x == "bmw":
    return int(2)
  elif x == "chevrolet":
    return int(3)
  elif x == "dodge":
    return int(4)
  elif x == "honda":
    return int(5)
  elif x == "isuzu":
    return int(6)
  elif x == "jaguar":
    return int(7) 
  elif x == "mazda":
    return int(8) 
  elif x == "mercedes-benz":
    return int(9) 
  elif x == "mitsubish":
    return int(10) 
  elif x == "nissan":
    return int(11) 
  elif x == "porsche":
    return int(12) 
  elif x == "toyota":
    return int(13)  
  elif x == "volkswagen":
    return int(14)  
  elif x == "volvo":
    return int(15) 

df2["Company"] = df2["company"].map(lambda x: company_map(x))
df2                                             
```
